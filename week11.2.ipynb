```python
import pandas as pd

df = pd.read_csv('/workspaces/DATA_3500_sandbox/NVDA_Stock.csv')
df.head()
```

```python
df['Close/Last'] = df['Close/Last'].replace({'\$':''}, regex=True).astype(float)
```

```python
df.iloc[::-1].reset_index(drop=True)
df
```

```python
# moving avg

df['MA_5'] = df['Close/Last'].shift(1).rolling(window=5).mean()
```

```python
buy = first_buy = total_profit = profit = 0

for i in range(len(df)):
    current_price = df.loc[i, 'Close/Last']
    moving_avg = df.loc[i, 'MA_5']

    if current_price < (moving_avg * 0.98):
        buy = current_price
        if first_buy == 0:
            first_buy = current_price
    elif current_price > (moving_avg * 1.02) and buy != 0:
        profit = current_price - buy
        total_profit += profit
        buy = 0
    else:
        df.loc[i, 'Signal'] = 'Sell'
```
