```python
"""
Programming Activity 1

Using the Titanic dataset and the Pandas library, fill in any missing data cells in the 'ages' column with the median age.
"""

dfT = pd.read_csv('/workspaces/DATA_3500_sandbox/titanic.csv')
median_age = dfT['Age'].median()
dfT['Age'].fillna(median_age, inplace=True)
dfT
```

```python
"""
Programming Activity 2

Using the Titanic dataset and the Pandas library, create a new column called 'FamilySize' and fill it with the size of the passenger's family.
"""

dfT['FamilySize'] = dfT['SibSp'] + dfT['Parch'] + 1
dfT
```

```python
"""
Programming Activity 3

Using the Titanic dataset and the Pandas library, find the average survival rate by class and sex.
"""

average_survival_rate = dfT.groupby(['Pclass', 'Sex'])['Survived'].mean().reset_index()
average_survival_rate
```

```python
"""
Programming Activity 6
Using the weather file (located on week 11 programming activities page), find the longest warm streak where the high temperature was above 65 degrees.

"""

dfW = pd.read_csv('/workspaces/DATA_3500_sandbox/SLC_weather.csv')
dfW['isWarm'] = dfW['Max Temp'] > 65
current_streak = 0
longest_streak = 0
for isWarm in dfW['isWarm']:
    if isWarm:
        current_streak += 1
        if current_streak > longest_streak:
            longest_streak = current_streak
    else:
        current_streak = 0

print(f"longest_streak: {longest_streak}")
```

```python
"""
Programming Activity 7

Using the same weather dataset, find the hottest day and the coldest night.
"""
hottest_day = dfW.loc[dfW['Max Temp'].idxmax()]
coldest_night = dfW.loc[dfW['Min Temp'].idxmin()]

print(f"hottest_day: {hottest_day}")
print(f"coldest_night: {coldest_night}")
```

```python
"""
Programming Activity 8
Using the same weather dataset, find the single most volatile day (the day with the biggest difference between high and low temperature).

"""

volatility = dfW['Max Temp'] - dfW['Min Temp']
most_volatile_day = dfW.loc[volatility.idxmax()]
print(f"most_volatile_day: {most_volatile_day}")
```
