```python
import seaborn as sns

import matplotlib.pyplot as plt

# summary of missing values
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
print(pd.concat([missing, missing_pct], axis=1, keys=['missing', 'percent']).loc[lambda x: x['missing'] > 0])

import seaborn.objects as so  # modern seaborn API

# assume df is your DataFrame
missing = df.isna().sum()

# --- Bar plot of missing counts ---
(
    so.Plot(
        data=pd.DataFrame({
            "column": missing[missing > 0].index,
            "missing": missing[missing > 0].values
        }),
        x="column", y="missing"
    )
    .add(so.Bar(), so.Agg())
    .scale(fill="viridis")  # replaces palette
    .label(y="Missing values", title="Missing values per column")
    .theme({
        "axes.labelrotation": 45,
        "axes.labelha": "right"
    })
    .layout(size=(8, 4))
    .show()
)

# --- Heatmap of missing pattern ---
plt.figure(figsize=(12, 3))
plt.imshow(df.isna().iloc[:200].T, aspect="auto", cmap="mako")
plt.xlabel("Rows (first 200)")
plt.title("Missing-data heatmap (columns vs rows)")
plt.yticks(range(len(df.columns)), df.columns, rotation=0)
plt.tight_layout()
plt.show()
```
